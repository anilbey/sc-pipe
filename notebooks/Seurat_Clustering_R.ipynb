```r
library(Seurat)
library(Matrix)
library(rhdf5)
```

```r
input_hdf5 = '../simulated/from_raw_counts/melanomaS2_sim_loc1_zheng17.h5'
h5f = H5Fopen(input_hdf5)
```

```r
matrix = h5f$matrix
H5Fclose(h5f)
```

```r
dim(matrix)
#genex vs cells
```

```r
matrix@dim_names = 5
```

```r
deget = as(matrix, "dgTMatrix")
```

```r
deget@Dimnames[[1]] = as.character((1:dim(matrix)[1]))
deget@Dimnames[[2]] = as.character((1:dim(matrix)[2]))
```

```r
deget@Dimnames
```

```r
as.character((1:dim(matrix)[1]))
```

```r
?CreateSeuratObject
```

### Seurat requires the dimnames of the DGT matrix

```r
seurat_object <- CreateSeuratObject(deget)
```

```r
#normalization of the gene expressions
seurat_object <- NormalizeData(object = seurat_object, normalization.method = "LogNormalize", 
                      scale.factor = 10000)
```

### Dimensionality reduction has to be computed prior to clustering

### Scaledata has to be performed before PCA

```r
seurat_object <- ScaleData(object = seurat_object)
```

```r
seurat_object <- RunPCA(object = seurat_object, pc.genes = seurat_object@var.genes, do.print = TRUE, pcs.print = 1:5, 
    genes.print = 5)
```

```r
seurat_object <- FindClusters(object = seurat_object, reduction.type = "pca", dims.use = 1:10, 
    resolution = 0.6, print.output = 0, save.SNN = FALSE)
```
