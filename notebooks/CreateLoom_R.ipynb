### To create loompy compatible files

```r
dummy_matrix = matrix(rexp(200, rate=.1), ncol=20)
dummy_row_attr = sample.int(100, 10)
dummy_col_attr = sample.int(100,20)
row_attr_list = list(cell_names = dummy_row_attr)
col_attr_list = list(gene_names = dummy_col_attr)
h5createFile("fakeloom.h5")
h5write(col_attr_list, "fakeloom.h5", "col_attrs" )
h5createGroup("fakeloom.h5","layers")
h5write(dummy_matrix, "fakeloom.h5","matrix")
h5write(row_attr_list, "fakeloom.h5", "row_attrs" )
```
